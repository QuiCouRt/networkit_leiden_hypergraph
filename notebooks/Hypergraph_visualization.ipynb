```python
import hypernetx as hnx
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
import warnings 
warnings.simplefilter('ignore')

from collections import defaultdict
```

```python
def read_hypergraph(file_path):
    scenes = {}
    with open(file_path, 'r') as file:
        for index, line in enumerate(file):
            scenes[index] = tuple(line.strip().split())
    return scenes

def read_partition(file_path):
    partition = []
    with open(file_path, 'r') as file:
        for line in file:
            partition.append(line.strip())
    return partition


file_path_hypergraph = '../output/Hypergraph.txt'
file_path_partition = '../output/Partition.txt'
scenes = read_hypergraph(file_path_hypergraph)
partition = read_partition(file_path_partition)

colors = defaultdict(lambda: plt.cm.tab10(len(colors)%10))

def get_node_color(v):
    return colors[partition[int(v)]]
    
H = hnx.Hypergraph(scenes)
hnx.drawing.draw(H, with_edge_labels=False, with_node_labels=False, nodes_kwargs={'color': get_node_color},edges_kwargs={'edgecolors': 'black'},)
```
